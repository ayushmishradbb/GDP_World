```python
fig, ax = plt.subplots(figsize=(16,6))

top_gdp_countries = df.sort_values('GDP ($ per capita)',ascending=False).head(20)
mean = pd.DataFrame({'Country':['World mean'], 'GDP ($ per capita)':[df['GDP ($ per capita)'].mean()]})
gdps = pd.concat([top_gdp_countries[['Country','GDP ($ per capita)']],mean],ignore_index=True)

sns.barplot(x='Country',y='GDP ($ per capita)',data=gdps, palette='Set3')
ax.set_xlabel(ax.get_xlabel(),labelpad=15)
ax.set_ylabel(ax.get_ylabel(),labelpad=30)
ax.xaxis.label.set_fontsize(16)
ax.yaxis.label.set_fontsize(16)
plt.xticks(rotation=90)
plt.show()
```

```python
for col in df.columns.values:
    if df[col].isnull().sum() == 0:
        continue
    if col == 'Climate':
        guess_values = df.groupby('Region')['Climate'].apply(lambda x: x.mode().max())
    else:
        guess_values = df.groupby('Region')[col].median()
    for region in df['Region'].unique():
        df[col].loc[(df[col].isnull())&(df['Region']==region)] = guess_values[region]
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(data=df.iloc[:,2:].corr(),annot=True,fmt='.2f',cmap='coolwarm')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df1 = df
```

```python
le = LabelEncoder()
df1['Region'] = le.fit_transform(df['Region'])
df1['Climate'] = le.fit_transform(df['Climate'])
```

```python
df1.head()
```

```python
df1 = df1.drop(['Country'], axis=1)
```

```python
df1.info()
```

```python
df1.corr()
```

```python
df1.corrwith(df1['GDP ($ per capita)'])
```

```python
X = df1.drop(['GDP ($ per capita)'], axis=1)
```

```python
y = df1['GDP ($ per capita)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)                                                                                                         
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()
model1.fit(X_train, y_train)
y_test_lr_pred = model1.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model2 = RandomForestRegressor(n_estimators = 50,
                             max_depth = 6,
                             min_weight_fraction_leaf = 0.05,
                             max_features = 0.8,
                             random_state = 42)
```

```python
model2.fit(X_train, y_train)
```

```python
y_test_rnd_pred = model2.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print("Linear Regression report:")
test_lr_mae = mean_absolute_error(y_test, y_test_lr_pred)
print("test_lr_mae:",test_lr_mae)
```

```python
print("Random Forest report:")
test_rnd_mae = mean_absolute_error(y_test, y_test_rnd_pred)
print("test_rnd_mae:",test_rnd_mae)
```
